```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
letter_thresh_minHSV = (int(230/2), 128, 0)
letter_thresh_maxHSV = (int(250/2), 255, 255)
```

```python
img1 = cv.imread("23.png")
plt.imshow(img1)
```

```python
# Try sharpening?
img2 = cv.medianBlur(img1, 11)
plt.imshow(img2)
```

```python
img3 = np.uint8(cv.Laplacian(cv.cvtColor(img1, cv.COLOR_BGR2GRAY), cv.CV_16S, 100))
plt.imshow(img3)
# Nope, not it
```

```python
img_HSV = cv.cvtColor(img1, cv.COLOR_BGR2HSV)
img_mask = cv.inRange(img_HSV, letter_thresh_minHSV, letter_thresh_maxHSV)
plt.imshow(img_mask)
```

```python
img_HSV = cv.cvtColor(img2, cv.COLOR_BGR2HSV)
img_mask = cv.inRange(img_HSV, letter_thresh_minHSV, letter_thresh_maxHSV)
plt.imshow(img_mask)
```
